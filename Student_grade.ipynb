```python
def add_student(name,age,Class):
  student = {
      'name': name,
      'age': age,
      'class':Class
  }
  return student

  students = {}

def add_grade(student, grade):
  if 'grade' not in student:
    student['grade'] = []
  student['grade'].append(grade)

  return students

def calculate_average(student):
  total = sum(student['grade'])
  average = total / len(student['grade'])
  return average

def grade_letter(average):
  if average >= 90:
    return 'A'
  elif average >= 80:
    return 'B'
  elif average >= 70:
    return 'C'
  elif average >= 60:
    return 'D'
  elif average >= 50:
    return 'E'
  else:
    return 'F'

def get_student_info(student):
  print(f"name: {student['name']}")
  print(f"age: {student['age']}")
  print(f"class: {student['class']}")
  print(f"grade: {student['grade']}")
  print(f"average: {student['average']}")
  print(f"grade_letter: {student['grade_letter']}")

  return students

def main():
  while True:
    print("1. Add student")
    print("2. Add grade")
    print("3. Calculate average")
    print("4. Get grade letter")
    print("5. Get student info")
    print("6. Exit")

    choice = input("Enter your choice: ")
    if choice == "1":
      name = input("Enter student name: ")
      age = input("Enter student age: ")
      Class = input("Enter student class: ")
      student = add_student(name,age,Class)
      student.update(student)
      print("student added successfully")

    elif choice == "2":
      student_name = input("Enter student name: ")
      grade = input("Enter grade:")
      for student in student:
        if student['name'] == student_name:
          add_grade(student,grade)
          print("grade added successfully")
          break
        else:
            print("student not found")

    elif choice == "3":
        student_name = input("Enter student name: ")
        for student in student:
          if student['name'] == student_name:
            average = calculate_average(student)
            print(f"average: {average}")
            break
          else:
              print("student not found")

    elif choice == "4":
      student_name = input("Enter student name: ")
      for student in student:
        if student['name'] == student_name:
          grade_letter = grade_letter(student['average'])
          print(f"grade_letter : {grade_letter}")
          break
        else:
          print("student not found")

    elif choice == "5":
      student_name = input("Enter student name:")
      for students in student:
        if student['name'] == student_name:
          get_student_info(student)
          break
        else:
          print('Student not found')

    elif choice == "6":
      print("Exiting program..")
      break

    else:
      print("invalid program try again!!")

main()
```
